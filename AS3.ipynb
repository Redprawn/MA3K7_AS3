```python
##Code 1: Calculate the final paper for a given 'n'
import numpy as np
from itertools import permutations
import random



def findfinal(n):
    Nlist = [i for i in range(1, n+1)]
    while len(Nlist) != 1:
        rand = random.sample(Nlist,2)
        Nlist.remove(rand[0])
        Nlist.remove(rand[1])
        new = abs(rand[0] - rand[1])
        Nlist.append(new)
    return Nlist[0]
```

```python
##Code 2: distribution of final papers graph ploting

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm
import seaborn as sns



splist = []
Musize = int(1000)
n = 2024
size = 100

for i in range(Musize):
    samples = findfinal(n)
    splist.append(samples)

    
fig,(ax1, ax2)=plt.subplots(1,2,figsize=(7,3))
x1 = np.linspace(1,2024)
pdf = uniform.pdf(x1)
ax1.plot(x1, pdf)
ax1.set_title(f'Uniform on [0,2024]') 
ax1.grid('on')

sns.histplot(splist, bins=25, 
             stat = 'density', ax=ax2)
ax2.set_title(f'Distribution of ') 
x2 = np.linspace(0, 1600)
ax2.set_ylabel('')
MU = np.mean(splist)
SIG = np.sqrt(np.var(splist, ddof=1))
y2 = norm.pdf(x2, MU, SIG) 
ax2.plot(x2,y2, 'r')
ax2.grid('on')

plt.show()
```
